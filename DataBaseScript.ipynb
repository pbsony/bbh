# [dbo].[sost]

```sql
/****** Object:  StoredProcedure [dbo].[sost]    Script Date: 20.02.2023 17:31:15 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[sost] @id int, @idk int, @kol int, @x int output
as begin 
declare @sum int
declare @p int 
declare @obsum int

select @p=Стоимость from Книга where ID=@idk
set @sum = @kol * @p
declare @old int 
select @old = ОбщаяСтоимость from Заказ where Номер=@id
set @obsum = @old + @sum



set @x=0
if @id not in (select @id from Заказ) 
set @x=1
if not exists (select ID from Книга where ID=@idk)
set @x=2
if exists (select * from СоставЗаказа where @id=НомерЗаказа and @idk=IDКниги )
begin 
set @x=4
declare @s int
select @s=КолвоКниг from СоставЗаказа where  IDКниги=@idk and @id=НомерЗаказа
set @s=@s+@kol
update СоставЗаказа set КолвоКниг=@s where IDКниги=@idk and @id=НомерЗаказа
update Заказ set ОбщаяСтоимость=@obsum where Номер=@id
end

if @x=0
begin 

insert into СоставЗаказа (НомерЗаказа, IDКниги, КолвоКниг) values (@id, @idk,@kol)
update Заказ set ОбщаяСтоимость=@obsum where Номер=@id
end 
end
GO
```
